# cleaning data



```python
# missing value

Data.isna().sum()
```

```python
# ather way

is_null = Data.isnull().sum().to_frame(name = 'is_null').T
is_na = Data.isna().sum().to_frame(name = 'is_na').T
unique = Data.nunique().to_frame(name = 'unique').T
pd.concat([is_na , is_null , unique])
```

we have not missing value

```python
# check duplicated

Data.duplicated().sum()
```

A lot of rows have the same values

```python
print(pd.unique(Data['Level']))
```

We value the level column numerically

```python
Data["Level"].replace({'High': 4, 'Medium': 3, 'Low': 2}, inplace=True)
Data
```

# Data plotting

```python
# noise

plt.figure(figsize=(15,30))
for i, col in enumerate(Data.columns):
    ax = plt.subplot(8, 3, i+1)
    sns.countplot(Data[col])
    plt.xticks(rotation = 90 , fontsize = 7)
    plt.xlabel(col)
        
plt.show()
```

According to the above plots, we do not have noise and outlying values

# 

```python
# Gender

sns.countplot(x = Data['Gender'] , data = Data , hue = 'Level')
plt.show()
```

We can see at what level men and women get cancer

```python
# Age

plt.figure(figsize=(20,7))
sns.countplot(x = Data['Age'] , data = Data)
plt.show()
```

At the age of 43, 51 , 54 and 63, there is less chance of getting this cancer

At the age of 33, 35 and 38, there is a higher chance of getting this cancer

    * It is better for people at this age to take necessary care for prevention and to perform necessary tests
    for timely diagnosis


People after the age of 63 get this cancer less often

People before the age of 14 get this cancer less

# 

```python
plt.figure(figsize=(15,30))
for i, col in enumerate(Data.columns):
    ax = plt.subplot(8, 3, i+1)
    Data.groupby(Data[col]).size().plot(kind='pie',  textprops={'fontsize': 6},autopct='%1.0f%%')
    
    plt.title(col)
        
plt.show() 
```

```python
plt.figure(figsize=(40,10))

Data.groupby(Data['Age']).size().plot(kind='pie',  textprops={'fontsize': 7},autopct='%1.0f%%')
    
plt.title('Age')
        
plt.show() 
```

Here, too, the results obtained for age groups are confirmed

# correlation

```python
fig , ax = plt.subplots(figsize = (20 , 15))
sns.heatmap(Data.corr() , cmap = 'RdBu_r' , cbar = True , annot = True , linewidths = 0.5 , ax = ax)
plt.show()
```

Obesity and coughing up blood have the greatest effect on the level of cancer and the column we want (target).
