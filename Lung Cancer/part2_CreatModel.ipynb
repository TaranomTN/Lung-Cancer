# importing library & Data

```python
# importing Data

import pandas as pd

# plot tree

import matplotlib.pyplot as plt

# creat matrix

import numpy as np

# creat model 

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import ComplementNB
from sklearn.neighbors import KNeighborsClassifier

# Model evaluation

from sklearn import metrics
from sklearn.metrics import classification_report , confusion_matrix
from sklearn.metrics import accuracy_score


from sklearn.model_selection import GridSearchCV

from sklearn import preprocessing     # normalize


# warning

import warnings
```

```python
warnings.simplefilter(action = 'ignore')
```

```python
# now importing the dataset

Data = pd.read_csv("cancer patient data sets.csv")
Data
```

```python
Data.drop(['index','Patient Id'] , axis = 1 , inplace = True)
Data
```

# creat model


# * decision tree

```python
# specify x and y

x = Data.drop('Level' , axis = 1)
y = Data['Level']
```

```python
# split x and y into training and testing set

x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.3 , random_state = 1)
```

```python
# call the algorithm

Classifier = DecisionTreeClassifier (max_depth = 3)


# fit the model with the training data

Classifier = Classifier.fit(x_train , y_train)


# predict with test data

y_pred = Classifier.predict(x_test)
```

# Model evaluation

```python
# Model score

print ("Accuracy :" , metrics.accuracy_score(y_test , y_pred))
```

```python
# Confusion matrix

confusion_matrix(y_test, y_pred)
```

```python
114 + 89 + 85
```

```python
# Evaluation for each class

print(classification_report (y_test , y_pred))
```

# Accuracy : 0.96

```python
# plot a decision tree


feature = ['Age', 'Gender', 'Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking',
       'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
       'Weight Loss', 'Shortness of Breath', 'Wheezing',
       'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold',
       'Dry Cough', 'Snoring']
target = ['Low' , 'Medium' , 'High' ]
fig = plt.figure(figsize = (25 ,20))
plot = tree.plot_tree(Classifier , feature_names = feature , class_names = target , filled = True )
```

As we concluded in part 1, coughing up blood has the greatest effect on the level of cancer

# Improve model

```python
# Using grid search to find the best parameters


grid_param={"criterion" : ["gini","entropy"] ,
             "splitter" : ["best","random"] ,
             "max_depth" : [2 , 3 , 5 , 10 , 20] ,
             "min_samples_leaf" : [2 , 5 ,10 , 20 , 50]}

grid_search = GridSearchCV( estimator = Classifier ,
                           param_grid = grid_param ,
                           scoring = "accuracy" ,
                           cv = 5 ,
                           n_jobs = -1 ,
                           verbose = 1 )

grid_search.fit(x_train,y_train)
```

```python
# Show the best parameters

print(grid_search.best_params_)
```

```python
# According to the best parameters

Classifier = DecisionTreeClassifier (
             criterion = 'gini',
             max_depth = 5 ,
             min_samples_leaf = 2 ,
             splitter='best' )


Classifier.fit(x_train , y_train)

y_pred = Classifier.predict(x_test)
```

# Model evaluation ( after improve model )

```python
# model score

print ("Accuracy :" , metrics.accuracy_score(y_test , y_pred))
```

With the improve model, we face the overfitting of the model !

# KNN

```python
# specify x and y

X = Data.drop('Level' , axis = 1)
Y = Data['Level']
```

```python
# split x and y into training and testing set

x_train , x_test , y_train , y_test = train_test_split( X , Y , test_size = 0.3 , random_state = 1)
```

# Finding the best k

```python
K = 20

# creat matrix of zeros with k dimention
Acc = np.zeros((K))
for i in range(1,K+1):
    clf = KNeighborsClassifier(n_neighbors = i)
    clf.fit(x_train,y_train.ravel())
    y_pred = clf.predict(x_test)
    Acc[i-1] = metrics.accuracy_score(y_test,y_pred)
    
Acc    
```

```python
# lowest and highest score

print(np.max(Acc))    # k = 1
print(np.min(Acc))    # k = 17
```

# plot

```python
training_acc = []
test_acc = []
neighbors_setting = range(1,20)
for n_neighbors in neighbors_setting:
    knn = KNeighborsClassifier(n_neighbors = n_neighbors)
    knn.fit(x_train,y_train.ravel())
    training_acc.append(knn.score(x_train,y_train))
    test_acc.append(knn.score(x_test,y_test))

plt.plot(neighbors_setting ,training_acc , label = "Accuracy of the training set" )
plt.plot(neighbors_setting ,test_acc , label = "Accuracy of the test set" )
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbors')
plt.grid()
plt.legend()
```

Shorter the distance is better
